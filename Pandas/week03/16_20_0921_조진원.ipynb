# 16. 과일로 만든 아이스크림 고르기

```python
import pandas as pd

# 데이터 load
df16_1 = pd.read_csv('../datasets/FIRST_HALF.csv')
df16_2 = pd.read_csv('../datasets/ICECREAM_INFO.csv')

# 데이터 합치기
df16 = pd.merge(df16_1, df16_2, how='inner', on='FLAVOR')

# 조건 뽑아내기
df16.loc[(df16['TOTAL_ORDER'] >= 3000) & (df16['INGREDIENT_TYPE'] == 'fruit_based')].sort_values(by=['TOTAL_ORDER'], ascending=False)['FLAVOR']
```

# 17. 평균 일일 대여 요금 구하기

```python
df17 = pd.read_csv('../datasets/CAR_RENTAL_COMPANY_CAR.csv', encoding='cp949')

average_fee = round(df17.loc[df17['car_type'] == 'SUV']['daily_fee'].mean())
pd.DataFrame([average_fee], columns=['AVERAGE_FEE'])
```

# 18. 특정 옵션이 포함된 자동차 리스트 구하기

# 19. Python 개발자 찾기

```python
df19 = pd.read_csv('../datasets/DEVELOPER_INFOS.csv')
df19[df19[['SKILL_1','SKILL_2','SKILL_3']].apply(lambda x : 'Python' in x.values, axis=1)].sort_values(by='ID', ascending=True)[['ID','EMAIL','FIRST_NAME','LAST_NAME']]
```

# 20. 조건에 부합하는 중고거래 댓글 조회하기

```python
df20_1 = pd.read_csv('../datasets/USED_GOODS_BOARD.csv', encoding='cp949')
df20_2 = pd.read_csv('../datasets/USED_GOODS_REPLY.csv', encoding='cp949')
df20 = pd.merge(df20_1, df20_2, how='right', on='BOARD_ID')

df20['CREATED_DATE_x'] = pd.to_datetime(df20['CREATED_DATE_x'], format='mixed')
df20['CREATED_DATE_y'] = pd.to_datetime(df20['CREATED_DATE_y'], format='mixed')

df20 = df20.loc[(df20['CREATED_DATE_x'].dt.year == 2022) & (df20['CREATED_DATE_x'].dt.month == 10)]
df20[['TITLE', 'BOARD_ID', 'REPLY_ID', 'CONTENTS_y', 'CREATED_DATE_y']].sort_values(by=['CREATED_DATE_y', 'TITLE'], ascending=[True,True]) 
```
