# 27. 자동차 대여 기록에서 장기/단기 대여 구분하기

```python
df27 = pd.read_csv('../datasets/CAR_RENTAL_COMPANY_RENTAL_HISTORY.csv')

df27['start_date'] = pd.to_datetime(df27['start_date'], format='mixed')
df27['end_date'] = pd.to_datetime(df27['end_date'], format='mixed')

import numpy as np
df27['RENT_TYPE'] = np.where((df27['end_date'] - df27['start_date']).dt.days >= 29, '장기 대여', '단기 대여')
df27.loc[(df27['start_date'].dt.year == 2022) & (df27['start_date'].dt.month == 9)].sort_values(by='history_id', ascending=False)
```

# 28. 조건에 맞는 도서 리스트 출력하기

```python
df28 = pd.read_csv('../datasets/BOOK.csv', encoding='cp949')

df28['published_date'] = pd.to_datetime(df28['published_date'], format='mixed')

df28.loc[(df28['category'] == '인문') & (df28['published_date'].dt.year == 2021)].sort_values(by='published_date', ascending=True)[['book_id','published_date']]
```
