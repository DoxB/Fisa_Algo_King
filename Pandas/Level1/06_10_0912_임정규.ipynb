### 00. 사전 데이터 로드

```python
import pandas as pd

animals_df = pd.read_csv('../datasets/ANIMAL_INS.CSV')
food_warehouse_df = pd.read_csv('../datasets/FOOD_WAREHOUSE.CSV', encoding='cp949')
food_factory_df = pd.read_csv('../datasets/FOOD_FACTORY.CSV', encoding='cp949')
```

### 06. 상위 n개 레코드

```python
ans = animals_df.sort_values('DATETIME')
ans[['NAME']]
```

### 07. 이름이 있는 동물의 아이디

```python
# dropna : 결측치를 제거하는 메소드

# 메소드 종류 : subset만 정의해주면 원하는 결과를 얻을 수 있다.

# ★ axis = 0 : NaN 값이 포함된 row를 drop (default)
# axis = 1 : NaN 값이 포함된 column을 drop

# ★ how =  ‘any’ : row 또는 column에 NaN 값이 1개만 있어도 drop(default)
# how =  ‘all’ : ow 또는 column에 있는 모든 값이 NaN이어야 drop

# ★ subset=[col1, col2, ...] : 해당 column에 대해서만 dropna를 진행

# inplace = True : dropna가 적용된 DataFrame 자체에 dropna를 적용
# ★ inplace = False : dropna가 적용된 DataFrame는 그대로 두고 dropna를 적용한 DataFrame을 return(default)

s_animals_df = animals_df.sort_values('ANIMAL_ID')
ans = s_animals_df.dropna(subset=['NAME'])
# ans[d_animals['NAME'].isna()]
ans[['NAME']]
```

### 08. 강원도에 위치한 생산공장 목록 출력하기

```python
# DataFrame 문자열 포함여부: df[].str.contains(조건)

s_food_factory_df = food_factory_df.sort_values('FACTORY_ID')
ans = s_food_factory_df[s_food_factory_df['ADDRESS'].str.contains('강원도')]
ans[['FACTORY_ID', 'FACTORY_NAME', 'ADDRESS']]
```

### 09. 최댓값 구하기

```python
# 열의 최소, 최대를 찾기 위해서는 .min, .max 사용 (axis 옵션 있음)

ans = animals_df[animals_df['DATETIME'] == animals_df['DATETIME'].min()]
ans[['DATETIME']]
```

### 10. 경기도에 위치한 식품창고 목록 출력하기

```python
# 결측치 채우기 메소드 fillna(문자열)

ans = food_warehouse_df
ans['FREEZER_YN'] = ans['FREEZER_YN'].fillna('N')
ans
```
