### 00. 사전 데이터 로드

```python
import pandas as pd
df = pd.read_csv('../datasets/ANIMAL_INS.CSV')
df
```

### 01. 모든 레코드 조회하기

```python
ans = df.sort_values('ANIMAL_ID')
ans
```

### 02. 역순 정렬하기

```python
s_df = df.sort_values('ANIMAL_ID', ascending=False)
ans = s_df[['NAME', 'DATETIME']]
```

### 03. 이름이 없는 동물의 아이디

```python
ans = df[df['NAME'].isna()][['ANIMAL_ID']]
ans
```

### 04. 동물의 아이디와 이름

```python
ans = df[['ANIMAL_ID', 'NAME']].sort_values('NAME')
ans
```

### 05. 여러 기준으로 정렬하기

```python
s_df = df[['ANIMAL_ID', 'NAME', 'DATETIME']]
ans = s_df.sort_values(by=['NAME', 'DATETIME'], ascending=[True, False])
ans
```
