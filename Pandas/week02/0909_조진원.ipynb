# 04. 동물의 아이디와 이름

# 05 여러 기준으로 정렬하기

```python
df = pd.read_csv('ANIMAL_INS.csv')

# 시계열 변환
df['DATETIME'] = pd.to_datetime(df['DATETIME'])

df = df.sort_values(by=['NAME','DATETIME'], ascending=[True,False], na_position='first')  
## na_position : 본래 결측값을 제외하고 나중에 정렬이 되지만, na_position을 하면 결측값의 위치를 지정할 수 있다.

df[['ANIMAL_ID','NAME','DATETIME']]
```
