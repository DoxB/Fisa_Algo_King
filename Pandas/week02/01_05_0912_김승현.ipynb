# 01. 모든 레코드 조회하기

```python
import pandas as pd

df = pd.read_csv("datasets/ANIMAL_INS.csv")

df1 = df.sort_values('ANIMAL_ID')
df1
```

# 02. 역순 정렬하기 모든 레코드 조회하기

```python
df1 = df.sort_values("ANIMAL_ID",ascending = False)
df1 = df1[['ANIMAL_ID','DATETIME']]
df1
```

# 03. 이름이 없는 동물의 아이디

```python
df1 = df[df.NAME.isna() == True][['ANIMAL_ID']]
df1
```

# 04. 동물 아이디와 이름

```python
df1 = df[['ANIMAL_ID','NAME']].sort_values('ANIMAL_ID')
df1
```

# 05.여러 기준으로 정렬하기

```python
df1 = df[['ANIMAL_ID','NAME','DATETIME']]
df2 = df1.sort_values(['NAME','DATETIME'], ascending = [True,False])
df2
```
